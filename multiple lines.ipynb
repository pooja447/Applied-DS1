```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('Max_Temp.csv')
data
```

```python
df=pd.DataFrame(data)
df
```

```python
days = list(df.iloc[:,0])
NewYork = list(df.iloc[:,1])
LosAngeles = list(df.iloc[:,2])
Philadelphia = list(df.iloc[:,3])
SanJose = list(df.iloc[:,4])
```

```python
plt.figure(figsize=(8,6))
plt.plot(days,NewYork, marker='o', label='NewYork')
plt.plot(days,LosAngeles, marker='o',label='LosAngeles')
plt.plot(days,Philadelphia,marker='o',label='Philadelphia')
plt.xlabel('Days')
plt.ylabel('Cities')
plt.legend()
plt.show()
```
